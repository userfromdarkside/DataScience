```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data['Price'] = data['Price'].str.replace('$','')
```

```python
data['Price'] = data['Price'].str.replace(',','')
```

```python
data['Price'] = data['Price'].str.replace('through-','')
```

```python
def calculate_mean(cell):
    if isinstance(cell, float):
        return cell  # Return the cell as it is if it's already a float
    try:
        numbers = cell.split()
        # Filter out invalid entries that cannot be converted to float
        valid_numbers = [num for num in numbers if num.replace('.', '', 1).isdigit()]
        if len(valid_numbers) == 2:
            return np.mean([float(num) for num in valid_numbers])
        else:
            return float(valid_numbers[0]) if valid_numbers else np.nan
    except (ValueError, IndexError):
        # Return NaN for invalid cells that cannot be processed
        return np.nan
```

```python
data['Price'] = data['Price'].apply(calculate_mean)
```

```python
import re
```

```python
def extract_rating(text):
    if isinstance(text,str):
        match = re.search(r'Rated (\d\.\d) out of 5 stars', text)
        return float(match.group(1)) if match else None
    return None
```

```python
data['Rating'] = data['Rating'].apply(extract_rating)
data.head()
```

```python
data.info()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(data['Rating'], data['Price'])
plt.xlabel('Rating')
plt.ylabel('Price')

plt.ylim(5,300)
plt.yticks(range(5,310,10))
```

```python
plt.figure(figsize=(60
                    ,6))
sns.violinplot(x='Sub Category', y='Rating', data=data, inner='quartile')
plt.title('Violin plot of rating by sub category')
plt.xlabel('Sub category')
plt.ylabel('Rating')
plt.ylim(1,5.6)
plt.yticks(range(1,6,1))
```
